```python
#1 O Reflorestador
listaDadosOriginais = []
meta = int (input('Digite sua meta de biomassa: '))
while True:
    bio = int (input('Digite o seu valor biomassa: '))
    listaDadosOriginais.append(bio)
    soma = sum(listaDadosOriginais)
    if soma > meta:
      print("\nLista de dados", listaDadosOriginais)
      print("Soma total da biomassa: ", soma)
      print("Árvores necessárias:",len(listaDadosOriginais))
      break
```

```python
#2 Simulador de combate
listaDanoMonster = []
listaDanoHero = []

while True:
    hero = int (input('Digite o dano que o monstro recebeu: '))
    listaDanoHero.append(hero)
    monster = int (input('Digite o dano que o herói recebeu: '))
    listaDanoMonster.append(monster)
    vidaH = 100 - sum(listaDanoMonster)
    vidaM = 100 - sum(listaDanoHero)
    print(f"\nVida do herói {vidaH}")
    print(f"Vida do monstro {vidaM}")

    if vidaH < 1 or vidaM < 1:
      print("\nCombate finalizado")
      break
    else:
      continue

```

```python
#3 Monitor de inundação
while True:
    nivel = int (input('Digite o nível do rio em metros: '))
    if nivel < 0:
      print("Programa encerrando...")
      break
    print(f"\nO nível do rio é de: {nivel} m")
    if nivel < 3:
           print("Estado Normal\n")
    elif nivel >= 3 and nivel <= 5:
          print("Estado de Alerta\n")
    else:
          print("Evacuação Imediata\n")
```

```python
#4 Coletor de itens
listaPeso = []
while True:
  item = float (input('Digite o peso do item em kg: '))
  listaPeso.append(item)
  soma = sum(listaPeso)
  exc = soma - 15
  if soma > 15:
    print(f"Mochila cheia, item descartado, pois o item excedeu em {exc:.2f} kg o limite de peso do seu inventário")
    break
  else:
    print(f"O peso atual do seu inventário é {soma}kg")
    continue


```

```python
#5 Média de temperatura global
temperaturas = []
dias = int(input("Digite quantos dias quer analisar: "))

i = 0
while i < dias:
    temp = float(input(f"Digite a temperatura do dia {i+1} em °C: "))
    temperaturas.append(temp)
    i += 1

if temperaturas:
    media = sum(temperaturas) / len(temperaturas)
    print(f"\nA média de temperatura durante esses dias é: {media:.2f}°C")
    if media > 25:
        print("Acima do esperado")
    else:
        print("Dentro da normalidade")
else:
    print("Nenhuma temperatura foi inserida.")
```
