```python
#PORCENTAGEM VITÓRIAS
print("PORCENTAGEM VITÓRIAS\n")
vitorias = int(input("Digite as vitórias do seu time:"))

empates = int(input("Digite os empates do seu time:"))

derrotas = int(input("Digite as derrotas do seu time:"))

pontos = (vitorias*3) + (empates*1) + (derrotas*0)

jogos = (vitorias + empates + derrotas)
print(f"\nSeu time jogou um total de: {jogos} rodadas")
print(f"Seu time fez um total de: {pontos} pontos ")

vitorias_porcentagem = (vitorias/jogos)*100
print(f"Seu time teve uma porcentagem de vitorias igual a: {vitorias_porcentagem:.2f}%\n")

if (vitorias_porcentagem) <= 35:
  print("ih rapaz, ta dando tudo errado, troca essas ideias ai em")

elif (vitorias_porcentagem) > 35 and vitorias_porcentagem <= 50:
  print("talvez considere trocar as táticas do seu time")

elif (vitorias_porcentagem) > 50 and vitorias_porcentagem <= 70:
  print("parabéns, seu time está bem no campeonato")

else:
  print("só pode ser o flamengo")


```

```python
#APROVEITAMENTO TOTAL
print ("APROVEITAMENTO TOTAL\n")

vitorias = int(input("Digite as vitórias do seu time:"))

empates = int(input("Digite os empates do seu time:"))

derrotas = int(input("Digite as derrotas do seu time:"))

pontos = (vitorias*3) + (empates*1) + (derrotas*0)

jogos = (vitorias + empates + derrotas)
print(f"\nSeu time jogou um total de:{jogos} rodadas")
print(f"Seu time fez um total de:{pontos} pontos ")

total = (jogos*3)

aproveitamento = ((pontos/total)*100)
print(f"\nSeu time teve um aproveitamendo de:{aproveitamento:.2f}%\n")

if (aproveitamento) <= 35:
  print("ih rapaz, ta dando tudo errado, troca essas ideias ai em")

elif (aproveitamento) > 35 and aproveitamento <= 50:
  print("talvez considere trocar as táticas do seu time")

elif (aproveitamento) > 50 and aproveitamento <= 70:
  print("parabéns, seu time está bem no campeonato")

else:
  print("só pode ser o flamengo")
```

```python
#ESTADOS DO BRASIL
estados = [
    "AC", "AL", "AP", "AM", "BA", "CE", "DF", "ES", "GO",
    "MA", "MT", "MS", "MG", "PA", "PB", "PR", "PE", "PI",
    "RJ", "RN", "RS", "RO", "RR", "SC", "SP", "SE", "TO"
]

while True:
  estado_digitado = input("Digite seu estado ").upper()

  if estado_digitado in estados:
    print(f"O estado '{estado_digitado}' é válido.")
    break
  else:
    print(f"O estado '{estado_digitado}' não é válido.")
```

```python
#ESTADOS DO BRASIL MANEIRA DIFERENTE
estados = [
    "AC", "AL", "AP", "AM", "BA", "CE", "DF", "ES", "GO",
    "MA", "MT", "MS", "MG", "PA", "PB", "PR", "PE", "PI",
    "RJ", "RN", "RS", "RO", "RR", "SC", "SP", "SE", "TO"
]

while True:
  estado_digitado = input("Digite seu estado ").upper()

  if estado_digitado not in estados:
    print(f"O estado '{estado_digitado}' não é válido.")

  else:
    print(f"O estado '{estado_digitado}' é válido.")
    break
```

```python
#EXEMPLO DE DATA DA AULA
from typing_extensions import final
# Implementando restrições
data = input('Digite uma data (DD/MM/AAAA):')
dia, mes, ano = data.split('/')
dia = int(dia)
mes = int(mes)
ano = int(ano)

if (dia < 1 or dia > 31):
  print('Dia inválido')
  exit()

if (mes < 1 or mes > 12):
  print('Mês inválido')
  exit()

if (ano < 1900 or ano > 2026):
  print('Ano inválido')
  exit()

print(f"sua data é {data}")
exit
```

```python
#CRIAR EMAIL
while True:
  nome_completo = input('Digite seu nome completo: ')
  palavras_nome = nome_completo.split()
  if (len(palavras_nome) >= 2):
    primeiro = palavras_nome[0].lower()
    sobrenome = palavras_nome[-1].lower()
    email = primeiro + '.' + sobrenome + '@ufn.edu.br'
    print('Seu email criado:', email)
    break

  print('Nome inválido')
```
