```python
#GLICOSE
momento = int (input('Digite a sua atual em mg/dl: '))
meta = int (input('Digite a sua meta em mg/dl : '))
sensi = int (input('Digite a sua sensibilidade (valor entre 20 a 60) : '))

remedio = ((momento - meta)/sensi)
print(f'Você precisa de:{remedio}ml/dl de insulina')
```

```python
#VIAGEM DE CARRO
distância = int (input('Digite a distância da viagem em km: '))
tempo = int (input('Digite a meta de horário em horas : '))


velocidade = (distância/tempo)
print(f'Sua velocidade média tem que ser de {velocidade}km/h')
```

```python
#PISCINA
altura = float (input('Digite a altura da piscina em metros: '))
lado1 = float (input('Digite um lado da piscina: '))
lado2 = float (input('Digite o outro lado da piscina: '))


volume = (lado1 * lado2 * altura)
print(f'Sua piscina precisa de {volume}L para encher')
```
