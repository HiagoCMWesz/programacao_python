```python
#9 GLICOSE E LISTA UM ABAIXO DO OUTRO
listaDadosOriginais = []
while True:
  glic = int (input('Digite o seu valor de glicemia (Digite 1 caso não queira mais registrar): '))
  if glic == 1:
    for glic in listaDadosOriginais:
      print(glic)
    print("\nFechando o programa...")
    break
  elif glic < 45 or glic > 450:
    print('Valor inválido, digite um valor entre 50 e 450')
  else: listaDadosOriginais.append(glic)
  continue



```

```python
#10 GLICOSE E LISTA ORDENADA
listaDadosOriginais = []
while True:
  glic = int (input('Digite o seu valor de glicemia (Digite 1 caso não queira mais registrar): '))
  if glic == 1:
    listaDadosOrdenados = listaDadosOriginais.copy()
    listaDadosOrdenados.sort()
    print("\nLista de dados", listaDadosOriginais)
    print("\nLista de dados em ordem crescente", listaDadosOrdenados)
    print("\nFechando o programa...")
    break
  elif glic < 45 or glic > 450:
    print('Valor inválido, digite um valor entre 50 e 450')
  else: listaDadosOriginais.append(glic)
  continue
```

```python
#11 GLICOSE MAIOR E MENOR VALOR
listaDadosOriginais = []
while True:
  glic = int (input('Digite o seu valor de glicemia (Digite 1 caso não queira mais registrar): '))
  if glic == 1:
    listaDadosOrdenados = listaDadosOriginais.copy()
    listaDadosOrdenados.sort()
    print("\nLista de dados", listaDadosOriginais)
    print("\nLista de dados em ordem crescente", listaDadosOrdenados)
    menor = listaDadosOrdenados[0]
    maior = listaDadosOrdenados[-1]
    print(f"\n{menor} foi o menor valor registrado de glicemia, enquanto {maior} foi o maior valor registrado de glicemia")
    print("\nFechando o programa...")
    break
  elif glic < 45 or glic > 450:
    print('Valor inválido, digite um valor entre 50 e 450')
  else: listaDadosOriginais.append(glic)
  continue
```

```python
#12 GLICOSE MEDIAS
listaDadosOriginais = []
while True:
  glic = int (input('Digite o seu valor de glicemia (Digite 1 caso não queira mais registrar): '))
  if glic == 1:
    if not listaDadosOriginais:
      print("Nenhum valor de glicemia foi registrado.")
      break

    print("\nLista de dados", listaDadosOriginais)

    media = sum(listaDadosOriginais) / len(listaDadosOriginais)
    print(f"\nA média de glicemia é {media:.2f}")

    valores_acima_media = [valor for valor in listaDadosOriginais if valor > media]
    valores_abaixo_media = [valor for valor in listaDadosOriginais if valor < media]

    if valores_acima_media:
      print(f"\nValores acima da média: {valores_acima_media}")
    else:
      print("\nNão há valores acima da média.")

    if valores_abaixo_media:
      print(f"\nValores abaixo da média: {valores_abaixo_media}")
    else:
      print("\nNão há valores abaixo da média.")

    print("\nFechando o programa...")
    break
  elif glic < 45 or glic > 450:
    print('Valor inválido, digite um valor entre 50 e 450')
  else: listaDadosOriginais.append(glic)
  continue
```
