```python
# Mostra todo Dataframe (lista)
print("LISTA:")
display(df.set_index('ID'))
```

```python
# Regras para qualificar o candidato
qualified_candidates = df[
    (df['Exp'] >= 2) &
    (df['Tecnico'] == 'Sim') &
    (df['Ingles'] == 'Sim') &
    (df['Idade'] >= 18) &
    (df['Idade'] <= 30)
]

# Mostrar os perfis adequados (ordem por ID)
print("\nPerfis Adequados (seguem todas as regras):")
display(qualified_candidates.set_index('ID'))
```

```python
# DESAFIO 4

import numpy as np

# Categorias Novas
df['Categoria'] = np.where(df['Exp'] > 5, 'SÊNIOR', 'JÚNIOR')

# Separar entre Júnior e Sênior
senior_candidates = df[df['Categoria'] == 'SÊNIOR']
junior_candidates = df[df['Categoria'] == 'JÚNIOR']

# Mostrar Sênior
print("\nCandidatos SÊNIOR:")
display(senior_candidates[['ID', 'Nome', 'Sobrenome', 'Exp', 'Tecnico', 'Ingles']].set_index('ID'))

# Mostrar Júnior
print("\nCandidatos JÚNIOR:")
display(junior_candidates[['ID', 'Nome', 'Sobrenome', 'Exp', 'Tecnico', 'Ingles']].set_index('ID'))
```
