```python
data = """Aragorn;Guerreiro;15;2;4500
Legolas;Arqueiro;22;1;5200
Gandalf;Mago;10;5;8000
Frodo;Ladrão;1;10;500
Gimli;Guerreiro;12;4;3800
Boromir;Guerreiro;8;8;2500
Galadriel;Mago;5;0;9500
Eowyn;Guerreiro;11;3;4100
Faramir;Arqueiro;9;4;3200
Saruman;Mago;18;2;8800
Sauron;Mago;30;1;15000
Gollum;Ladrão;2;15;300
Sam;Ladrão;4;2;1200
Merry;Ladrão;3;4;800
Pippin;Ladrão;2;5;700
Elrond;Mago;7;1;6000
Arwen;Mago;4;0;4500
Theoden;Guerreiro;10;6;3500
Ugluk;Guerreiro;6;12;2000
Lurtz;Arqueiro;14;5;4800"""

with open('partida.txt', 'w') as f:
    f.write(data)

print("File 'partida.txt' created successfully.")


def ler_e_tratar_dados(caminho_arquivo):
    jogadores = []
    with open(caminho_arquivo, 'r') as f:
        for linha in f:
            dados = linha.strip().split(';')
            if len(dados) == 5:
                jogador = {
                    "nome": dados[0],
                    "classe": dados[1],
                    "kills": int(dados[2]),
                    "deaths": int(dados[3]),
                    "dano": int(dados[4])
                }
                jogadores.append(jogador)
    return jogadores

jogadores_data = ler_e_tratar_dados('partida.txt')
print("Dados tratados dos jogadores:")
for jogador in jogadores_data:
    print(jogador)
```

```python
def calcular_kda(kills, deaths):
    if deaths == 0:
        return float(kills)
    return kills / deaths

print(f"KDA de Aragorn (15 kills, 2 deaths): {calcular_kda(15, 2):.2f}")
print(f"KDA de Galadriel (5 kills, 0 deaths): {calcular_kda(5, 0):.2f}")
```

```python
def filtrar_por_classe(jogadores, classe):
    return [jogador for jogador in jogadores if jogador['classe'].lower() == classe.lower()]


magos = filtrar_por_classe(jogadores_data, 'Mago')
print("\nJogadores da classe 'Mago':")
for mago in magos:
    print(mago['nome'])
```

```python
maior_dano = {'nome': '', 'dano': -1}
for jogador in jogadores_data:
    if jogador['dano'] > maior_dano['dano']:
        maior_dano['nome'] = jogador['nome']
        maior_dano['dano'] = jogador['dano']
print(f"\nJogador com maior dano: {maior_dano['nome']} ({maior_dano['dano']} de dano)")

total_kills = sum(jogador['kills'] for jogador in jogadores_data)
media_kills = total_kills / len(jogadores_data)
print(f"Média de kills da partida: {media_kills:.2f}")

jogadores_kda_alto = []
for jogador in jogadores_data:
    kda = calcular_kda(jogador['kills'], jogador['deaths'])
    if kda > 2.0:
        jogadores_kda_alto.append(jogador['nome'].upper())
print(f"Jogadores com KDA superior a 2.0: {', '.join(jogadores_kda_alto)}")
```
